```python
#importing required modules
import pandas as pd
import numpy as np
```

###  I have Loaded the data from wikipedia to excel and then imported data here in Jupyter Lab

```python
# Importing Data in Lab
df = pd.read_excel('project data 3.xlsx')
df.head()
```

```python
# Replacing values with Not null values.
df.replace("Not assigned",np.nan,inplace=True)
df.head()
```

```python
# Droping columns with null values
df.dropna(subset=["Borough"], axis=0 ,inplace=True)
```

```python
# Visualizing dataframe
df.head(15)
```

```python
# reseting the index
df.reset_index(drop=True,inplace=True)
df
```

```python
# Checkingg if the further values are null
df.isnull()
```

```python
#Counting the missing values
missing_data=df.isnull()
for columns in missing_data:
    print(columns)
    print(missing_data[columns].value_counts())
    print(" ")
```
